Задача 2. Написать функцию для вычисления факториала и рассчитать число сочетаний из 6 по 2.

```python
def fact(x):
    if x == 0:
        return 1
    return fact(x - 1) * x
```

Задача 3. Из 30 стрелков 12 попадает в цель с вероятностью 0,6, 8 стрелков — с вероятностью 0,5, а 10 — с вероятностью 0,7. Наудачу выбранный стрелок произвёл выстрел, поразив цель.
К какой из групп, вероятнее всего, принадлежал этот стрелок?

Событие ${B-n}$ спортсмен принадлежит группе n.
Вероятность выстрела спортсменом
$P(B_1) = \frac 2 5,~ P(B_2) = \frac 4 {15},~ P(B_3) = \frac 1 3$

Событие ${A-}$ спортсмен попал. Условные вероятности попаданий:

$P(A|B_{1}) = 0.6, P(A|B_{2}) = 0.5, P(A|B_{3}) = 0.7$

Найдем полную вероятность:
$$P(A)=P(B_{1})\cdot P(A|B_{1})+P(B_{2})\cdot P(A|B_{2})+ \dots +P(B_{n})\cdot P(A|B_{n})$$
$$P(A) = 2/5 \cdot 0.6 + 4/15 \cdot 0.5 + 1/3 \cdot 0.7 = 0.606666 = 0.607$$

Согласно теореме Байеса:

a) вероятность того, что выстрел произведен спортсменом 1 группы
$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)} = \frac{2}{5} \cdot \frac{0.6}{0.607} = 0.395 $$

б) спортсменом 2 группы
$$P(B_2\:|\:A) = \frac{P(B_2) \cdot P(A\:|\:B_2)}{P(A)} = \frac{4}{15} \cdot \frac{0.5}{0.607} = 0.22 $$

в) спортсменом 3 группы
$$P(B_3\:|\:A) = \frac{P(B_3) \cdot P(A\:|\:B_3)}{P(A)} = \frac{1}{3} \cdot \frac{0.7}{0.607} = 0.384 $$

Проверка: $0,395 + 0,22 + 0,384 = 1$

Ответ: Вероятнее всего, стрелок принадлежал 1 группе

Задача 4. Сгенерировать выборку реализаций случайной величины (кроме распределения Коши).
Любимое распределение можно выбрать на этом [сайте](https://docs.scipy.org/doc/scipy/reference/stats.html).
Проверить ЦПТ для различных n.

```python
from scipy.stats import expon
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
lam = .1
expon_rv = sps.expon(scale=1 / lam)
# генерация выборки
rvs_size = 1000
sns.distplot(expon_rv.rvs(rvs_size), bins=10)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 20))
for num, n in enumerate([5, 10, 50, 200]):
    a1 = []
    for _ in range(rvs_size):
        a1.append(expon_rv.rvs(size=n).mean())
    i, j = (num // 2, num % 2)

    means = [np.mean(expon_rv.rvs(n)) for _ in range(rvs_size)]

    mu = 1 / lam  # theoretical expectation
    mu_means = np.mean(means)

    variance = 1 / lam ** 2
    sigma = (variance / n) ** 0.5  # theoretical sigma
    sigma_test = np.std(means, ddof=0)  # theoretical sigma
    sigma_means = np.std(means, ddof=1)

    print("Матожидание выборки объема %d: %.4f" % (n, mu_means))
    print("Матожидание НР: %.4f" % mu)
    print("Среднее квадратичное отклонение выборки объема %d: %.4f" % (n, sigma_means))
    print("Среднее квадратичное отклонение НР: %.4f" % sigma)

    sns.distplot(a1, bins=10, ax=axs[i, j])
    axs[i, j].set_title(f'n = {n}', fontsize=20)
    axs[i, j].set_xlabel('$x$', fontsize=18)
    axs[i, j].set_ylabel('density', fontsize=18)
    axs[i, j].legend(['Гистограмма распред-й\nвыборочных средних'], fontsize=15)
    print(sps.anderson(a1))
```

Эксперимент показал, гистограмма сгенерированных выборок близко аппроксимирует график нормального распределения, построенный с использованием центральной предельной теоремы. Значения математических ожиданий и стандартных отклонений близки и практически совпадают.

Эксперимент подтвердил теорию тем, что для заданного распределения центральная предельная теорема работает с высокой точностью и при небольшом размере выборки.

Задача 5. Попробовать применить ЦПТ к распределению Коши.

```python
cauchy_rv = sps.cauchy()
fig, axs = plt.subplots(2, 2, figsize=(20, 20))
for num, n in enumerate([5, 10, 50, 200]):
    a2 = []
    for _ in range(rvs_size):
        a2.append(cauchy_rv.rvs(size=n).mean())
    i, j = (num // 2, num % 2)

    sns.distplot(a2, bins=10, ax=axs[i, j])
    axs[i, j].set_title(f'n = {n}', fontsize=20)
    axs[i, j].set_xlabel('$x$', fontsize=18)
    axs[i, j].set_ylabel('density', fontsize=18)
    axs[i, j].legend(['Гистограмма распред-й\nвыборочных средних'], fontsize=15)
    print(sps.anderson(a2))
```
